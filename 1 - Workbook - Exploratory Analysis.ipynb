## I. Basic Information

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("real_estate_data.csv")
df.shape
df.dtypes
```

```python
df.head()
```

## II. Distributions

```python
df.hist()
plt.show()
```

```python
plt.show()
```

```python
df.year_built.hist()
plt.show()
df.describe()
```

```python
# Summarize categorical features
df.describe(include=['object'])
```

```python
# Bar plot for 'exterior_walls'
sns.countplot(y='exterior_walls', data=df)
plt.show()
```

## III. Segmentations

```python
# Segment tx_price by property_type and plot distributions
sns.boxplot(y='property_type', x='tx_price', data=df)
```

```python
# Segment by property_type and display the means within each class
df.groupby('property_type').mean(numeric_only=True)
```

## IV. Correlations

```python
# Calculate correlations between numeric features
correlations = df.corr(numeric_only=True)
correlations
```

```python
# Increase the figsize to 7 x 6
plt.figure(figsize=(7,6))

# Plot heatmap of correlations
sns.heatmap(correlations, cmap='RdBu_r')
```
